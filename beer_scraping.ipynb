```python
from splinter import Browser
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import pandas as pd
```

```python
req = Request(
    url='https://www.beeradvocate.com/beer/top-rated/us/mi/', 
    headers={'User-Agent': 'Chrome/5.0'}
)
webpage = urlopen(req).read()
print(webpage)
```

```python
# Scrape the website

# Create a BeautifulSoup object from the scraped HTML
soup = BeautifulSoup(webpage, 'html.parser')
```

```python
ba_element = soup.find_all('td', class_ = 'hr_bottom_light')
ba_element
```

```python
beer_list = []

for row in ba_element:
    data = row.get_text()
    print(data)
```

```python
beer_list
```

```python
column_names = ["beer name"]
beer_df = pd.DataFrame(beer_list, columns=column_names)
beer_df
```
